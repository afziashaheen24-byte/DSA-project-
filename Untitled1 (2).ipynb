```python
import tkinter as tk
from tkinter import messagebox, simpledialog
import webbrowser
import json
import os

# --- Data Models ---
UNIVERSITY_OPTIONS = {
    "Pre_Med": {
        "High (80-100)": ["MBBS", "BDS", "BS Pharmacy"],
        "Medium (70-79)": ["DPT (Physiotherapy)", "BS Generic Nursing", "BS Biotechnology"],
        "Low (60-69)": ["BS Medical Lab Technology (MLT)", "BS Food Science & Technology", "Paramedic Programs"]
    },
    "Pre_Eng": {
        "High (80-100)": ["BS Software Engineering", "BS Electrical Engineering", "BS Computer Engineering"],
        "Medium (70-79)": ["BS Civil Engineering", "BS Mechanical Engineering", "BS City & Regional Planning"],
        "Low (60-69)": ["BS Technology Programs", "Associate Degree Programs (ADP)", "BSc Mathematics/Physics"]
    },
    "ICS": {
        "High (80-100)": ["BS Computer Science (Top Unis)", "BS Data Science", "BS Artificial Intelligence"],
        "Medium (70-79)": ["BS Information Technology", "BS Software Engineering (General)", "BS Web Development"],
        "Low (60-69)": ["BS Digital Marketing", "BS Commerce/Accounting", "Associate Degree in IT"]
    },
    "ICom": {
        "High (80-100)": ["BBA (Top Programs)", "BS Accounting & Finance", "CA Foundation"],
        "Medium (70-79)": ["BBA (General)", "BS Banking & Finance", "BS Supply Chain Management"],
        "Low (60-69)": ["BS Commerce", "Associate Degree in Accounting", "Diploma in Business Admin"]
    },
    "FA": {
        "High (80-100)": ["LLB (Law)", "BS International Relations", "BS Psychology"],
        "Medium (70-79)": ["BS Mass Communication", "BS English Literature", "BS Sociology"],
        "Low (60-69)": ["BS Political Science", "Associate Degree in Arts", "Diploma in Journalism"]
    }
}

APTITUDE_QUESTIONS = [
    ("Do you enjoy solving complex math problems and puzzles?", "Logic"),
    ("Are you curious about how living organisms and the human body work?", "Biology"),
    ("Do you like to manage money, understand markets, and think about business strategies?", "Business"),
    ("Are you interested in history, literature, languages, or understanding human behavior?", "Creative/Social"),
    ("Do you like to build things or understand machines and systems?", "Logic"),
    ("Do experiments related to biology or chemistry excite you?", "Biology"),
    ("Are you good at organizing, planning, and keeping records?", "Business"),
    ("Do you express yourself well through writing or public speaking?", "Creative/Social")
]

INTERMEDIATE_PATHS_INFO = {
    "Pre-Engineering (FSc)": {
        "subjects": "Physics, Chemistry, Mathematics",
        "description": "Best for students interested in Engineering, Technology and Problem Solving.\n\nFuture Scope:\n• Electrical, Mechanical, Civil, Software Engineering\n• Architecture, Robotics, Mechatronics\n• High-paying govt and private engineering jobs\n• Great scope in overseas jobs (UAE, USA, UK)"
    },
    "Pre-Medical (FSc)": {
        "subjects": "Physics, Chemistry, Biology",
        "description": "Ideal for students who want to enter the medical field.\n\nFuture Scope:\n• MBBS, BDS, DPT, Nursing, Pharmacy\n• Biotechnology, Medical Lab Technology\n• High-respect and high-scope fields in Pakistan and abroad\n• Great future in research, hospitals, and public health"
    },
    "ICS (Computer Science)": {
        "subjects": "Computer Science, Mathematics, Physics",
        "description": "Perfect for students interested in IT, Software, and Technology.\n\nFuture Scope:\n• BSCS, BS SE, AI, Data Science\n• Cybersecurity, Cloud Computing, Gaming\n• Remote international jobs (USA, Europe) without leaving Pakistan\n• Very high salary potential"
    },
    "ICom (Commerce)": {
        "subjects": "Accounting, Economics, Commerce",
        "description": "Best for students planning for Business, Finance, and Management careers.\n\nFuture Scope:\n• BBA, Accounting & Finance, CA, ACCA\n• Banking, HR, Marketing, Supply Chain\n• Own business, entrepreneurship, corporate jobs"
    },
    "FA (Arts/Humanities)": {
        "subjects": "Social Sciences, Languages, Arts",
        "description": "Ideal for creative and social fields.\n\nFuture Scope:\n• Law, Psychology, Mass Communication\n• IR, Sociology, English Literature\n• Journalism, Civil Services, Teaching\n• Good for students who want non-technical fields"
    }
}


# --- Resource Management (new) ---
RESOURCES_FILE = "resources.json"
DEFAULT_RESOURCES = [
    {
        "title": "Khan Academy - Biology",
        "type": "Study Material",
        "category": "Biology",
        "description": "Free video lessons and practice exercises for biology basics.",
        "url": "https://www.khanacademy.org/science/biology"
    },
    {
        "title": "edX - CS50 (Intro to CS)",
        "type": "Online Course",
        "category": "Computer Science",
        "description": "Harvard's introductory CS course (edX).",
        "url": "https://www.edx.org/course/cs50s-introduction-to-computer-science"
    },
    {
        "title": "Punjab University - Official",
        "type": "University Link",
        "category": "Top Universities",
        "description": "Official Punjab University website for admissions and programs.",
        "url": "http://pu.edu.pk"
    },
    {
        "title": "Coursera - Data Science Specialization",
        "type": "Online Course",
        "category": "Data Science",
        "description": "Coursera specialization covering Python, Statistics and Machine Learning.",
        "url": "https://www.coursera.org/specializations/jhu-data-science"
    }
]


def load_resources():
    if os.path.exists(RESOURCES_FILE):
        try:
            with open(RESOURCES_FILE, "r", encoding="utf-8") as f:
                return json.load(f)
        except Exception:
            return DEFAULT_RESOURCES.copy()
    else:
        # create default file
        try:
            with open(RESOURCES_FILE, "w", encoding="utf-8") as f:
                json.dump(DEFAULT_RESOURCES, f, indent=2, ensure_ascii=False)
        except Exception:
            pass
        return DEFAULT_RESOURCES.copy()


def save_resources(resources):
    try:
        with open(RESOURCES_FILE, "w", encoding="utf-8") as f:
            json.dump(resources, f, indent=2, ensure_ascii=False)
    except Exception as e:
        print("Could not save resources:", e)


# --- Main Application ---
class SkillRouteAnalyzer:
    def __init__(self, master):
        self.master = master
        master.title("Skill Route Analyzer")
        master.geometry("1000x700")
        master.config(bg="#E8F0F8")
        self.logged_in = False

        # load resources
        self.resources = load_resources()

        # Header
        self.header_frame = tk.Frame(master, bg="#1F1F2E", height=80)
        self.header_frame.pack(fill='x', side='top')
        tk.Label(self.header_frame, text="Skill Route Analyzer", font=("Arial", 28, "bold"),
                 bg="#1F1F2E", fg="white").pack(side='left', padx=20, pady=15)
        self.logout_btn = tk.Button(self.header_frame, text="Login", command=self.toggle_login,
                                    font=("Arial", 12), bg="#1F1F2E", fg="white", bd=0)
        self.logout_btn.pack(side='right', padx=20)

        # Content
        self.content_frame = tk.Frame(master, bg="#E8F0F8")
        self.content_frame.pack(fill='both', expand=True, padx=20, pady=20)

        self.show_login_screen()

    def toggle_login(self):
        if self.logged_in:
            self.logged_in = False
            self.logout_btn.config(text="Login")
            self.show_login_screen()
        else:
            self.logged_in = True
            self.logout_btn.config(text="Logout")
            self.show_dashboard()

    def clear_content_frame(self):
        for widget in self.content_frame.winfo_children():
            widget.destroy()

    # --- Login ---
    def show_login_screen(self):
        self.clear_content_frame()
        tk.Label(self.content_frame, text="Login to Continue", font=("Arial", 24, "bold"),
                 bg="#E8F0F8", fg="#6C63FF").pack(pady=50)
        frame = tk.Frame(self.content_frame, bg="#F0F4F8", bd=1, relief="solid", padx=40, pady=30)
        frame.pack(pady=20)
        tk.Label(frame, text="Username:", font=("Arial", 14), bg="#F0F4F8").pack(anchor='w')
        self.username_entry = tk.Entry(frame, font=("Arial", 14), width=25)
        self.username_entry.pack(pady=5)
        tk.Label(frame, text="Password:", font=("Arial", 14), bg="#F0F4F8").pack(anchor='w')
        self.password_entry = tk.Entry(frame, font=("Arial", 14), show='*', width=25)
        self.password_entry.pack(pady=5)
        tk.Button(frame, text="Login", font=("Arial", 14, "bold"), bg="#6C63FF", fg="white", width=20,
                  command=self.login_action).pack(pady=20)

    def login_action(self):
        # Accept any non-empty username/password
        if self.username_entry.get() != "" and self.password_entry.get() != "":
            self.logged_in = True
            self.logout_btn.config(text="Logout")
            self.show_dashboard()
        else:
            messagebox.showerror("Error", "Username or password cannot be empty")

    # --- Dashboard ---
    def show_dashboard(self):
        if not self.logged_in:
            self.show_login_screen()
            return

        self.clear_content_frame()
        tk.Label(self.content_frame, text="Welcome!", font=("Arial", 22, "bold"), bg="#E8F0F8").pack(pady=20)

        tk.Button(self.content_frame, text="1. Start Stream Assessment (Grade 9)", width=40, height=2,
                  bg="#6C63FF", fg="white", command=self.show_quiz).pack(pady=10)

        tk.Button(self.content_frame, text="2. Explore Intermediate Paths", width=40, height=2,
                  bg="#6C63FF", fg="white", command=self.show_path_info).pack(pady=10)

        tk.Button(self.content_frame, text="3. University Percentage Guide", width=40, height=2,
                  bg="#6C63FF", fg="white", command=self.show_percentage_guide_input).pack(pady=10)

        # NEW Resource Management button
        tk.Button(self.content_frame, text="4. Resources & Courses (Study Material, Online Courses, University Links)",
                  width=60, height=2, bg="#4CAF50", fg="white",
                  command=self.show_resources_screen).pack(pady=10)

    # --- Quiz ---
    def show_quiz(self):
        self.clear_content_frame()
        self.quiz_scores = {"Logic": 0, "Biology": 0, "Business": 0, "Creative/Social": 0}
        self.question_index = 0

        tk.Label(self.content_frame, text="Stream Selection Quiz (Grade 9)", font=("Arial", 20, "bold"),
                 bg="#E8F0F8").pack(pady=20)

        self.question_label = tk.Label(self.content_frame, text="", font=("Arial", 14),
                                       bg="#E8F0F8", wraplength=700, justify='left')
        self.question_label.pack(pady=20)

        self.options_var = tk.StringVar(value="")  # Required fix

        for option in ["Strongly Agree", "Agree", "Neutral", "Disagree", "Strongly Disagree"]:
            tk.Radiobutton(self.content_frame, text=option, variable=self.options_var,
                           value=option, bg="#E8F0F8").pack(anchor='w')

        tk.Button(self.content_frame, text="Next Question", bg="#FF6584",
                  fg="white", command=self.next_question).pack(pady=10)

        tk.Button(self.content_frame, text="Back to Dashboard", bg="#FF6584",
                  fg="white", command=self.show_dashboard).pack(pady=5)

        self.load_question()

    def load_question(self):
        if self.question_index < len(APTITUDE_QUESTIONS):
            q, _ = APTITUDE_QUESTIONS[self.question_index]
            self.question_label.config(text=f"Q{self.question_index + 1}: {q}")
            self.options_var.set("")  # clear selection
        else:
            self.calculate_quiz_result()

    def next_question(self):
        if not self.options_var.get():
            tk.messagebox.showwarning("Selection Required",
                                      "Please select an option before proceeding!")
            return

        _, cat = APTITUDE_QUESTIONS[self.question_index]
        score_map = {"Strongly Agree": 2, "Agree": 1, "Neutral": 0, "Disagree": -1, "Strongly Disagree": -2}
        self.quiz_scores[cat] += score_map[self.options_var.get()]

        self.question_index += 1
        self.load_question()

    def calculate_quiz_result(self):
        best = max(self.quiz_scores, key=self.quiz_scores.get)
        result = {
            "Logic": "Pre-Engineering or ICS",
            "Biology": "Pre-Medical",
            "Business": "ICom",
            "Creative/Social": "FA (Arts)"
        }.get(best, "Mixed Results")

        tk.messagebox.showinfo("Quiz Result", f"Recommended Stream: {result}")
        self.show_dashboard()

    # --- Intermediate Paths ---
    def show_path_info(self):
        self.clear_content_frame()

        tk.Label(self.content_frame, text="Explore Intermediate Paths",
                 font=("Arial", 20, "bold"), bg="#E8F0F8").pack(pady=20)

        # Scrollable Frame
        main_frame = tk.Frame(self.content_frame, bg="#F0F4F8", bd=1, relief="solid")
        main_frame.pack(fill="both", expand=True)

        canvas = tk.Canvas(main_frame, bg="#F0F4F8")
        scrollbar = tk.Scrollbar(main_frame, orient="vertical", command=canvas.yview)
        scroll_frame = tk.Frame(canvas, bg="#F0F4F8")

        scroll_frame.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
        canvas.create_window((0, 0), window=scroll_frame, anchor="nw")
        canvas.configure(yscrollcommand=scrollbar.set)

        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")

        # Fill data
        for path, data in INTERMEDIATE_PATHS_INFO.items():
            tk.Label(scroll_frame, text=f"▶ {path}", font=("Arial", 16, "bold"),
                     bg="#F0F4F8", fg="#6C63FF").pack(anchor='w', pady=(20, 0))

            tk.Label(scroll_frame, text=f"Subjects: {data['subjects']}",
                     bg="#F0F4F8", fg="black", font=("Arial", 12, "italic")).pack(anchor='w', padx=20)

            tk.Label(scroll_frame, text=data['description'], bg="#F0F4F8",
                     wraplength=900, justify="left", font=("Arial", 12)).pack(anchor="w", padx=20, pady=5)

        # BACK button
        tk.Button(scroll_frame, text="Back to Dashboard", bg="#FF6584", fg="white",
                  command=self.show_dashboard).pack(pady=20)

    # --- University Percentage ---
    def show_percentage_guide_input(self):
        self.clear_content_frame()

        tk.Label(self.content_frame, text="University Percentage Guide",
                 font=("Arial", 20, "bold"), bg="#E8F0F8").pack(pady=20)

        frame = tk.Frame(self.content_frame, bg="#F0F4F8",
                         bd=1, relief="solid", padx=20, pady=20)
        frame.pack(pady=10)

        tk.Label(frame, text="Select Stream:", bg="#F0F4F8").pack()
        self.stream_var = tk.StringVar(value="Pre_Med")
        tk.OptionMenu(frame, self.stream_var, *UNIVERSITY_OPTIONS.keys()).pack(pady=5)

        tk.Label(frame, text="Enter Percentage:", bg="#F0F4F8").pack()
        self.perc_entry = tk.Entry(frame)
        self.perc_entry.pack(pady=5)

        tk.Button(frame, text="Show Options", bg="#FF6584", fg="white",
                  command=self.show_percentage_results).pack(pady=10)

        tk.Button(self.content_frame, text="Back to Dashboard", bg="#FF6584",
                  fg="white", command=self.show_dashboard).pack(pady=10)

    def show_percentage_results(self):
        try:
            perc = float(self.perc_entry.get())
        except:
            tk.messagebox.showerror("Error", "Invalid percentage")
            return

        stream = self.stream_var.get()

        if perc >= 80:
            cat = "High (80-100)"
        elif perc >= 70:
            cat = "Medium (70-79)"
        elif perc >= 60:
            cat = "Low (60-69)"
        else:
            tk.messagebox.showwarning("Low", "Percentage below 60%")
            return

        opts = UNIVERSITY_OPTIONS[stream][cat]
        tk.messagebox.showinfo("Options",
                               f"Stream: {stream}\nPercentage: {perc}%\n\nDegrees:\n" + "\n".join(opts))

    # --- Resources Screen (NEW) ---
    def show_resources_screen(self):
        self.clear_content_frame()
        tk.Label(self.content_frame, text="Resources & Courses", font=("Arial", 20, "bold"),
                 bg="#E8F0F8").pack(pady=10)

        control_frame = tk.Frame(self.content_frame, bg="#E8F0F8")
        control_frame.pack(fill="x", padx=10, pady=5)

        tk.Label(control_frame, text="Search:", bg="#E8F0F8").pack(side="left", padx=(0, 5))
        self.res_search_var = tk.StringVar()
        tk.Entry(control_frame, textvariable=self.res_search_var, width=40).pack(side="left", padx=(0, 10))
        tk.Button(control_frame, text="Search", command=self.filter_resources, bg="#6C63FF", fg="white").pack(side="left")

        tk.Button(control_frame, text="Show All", command=self.show_all_resources, bg="#FF6584", fg="white").pack(side="left", padx=10)

        # Add Resource (admin) and Back
        tk.Button(control_frame, text="Add Resource", command=self.add_resource_dialog, bg="#4CAF50", fg="white").pack(side="right")
        tk.Button(control_frame, text="Back", command=self.show_dashboard, bg="#FF6584", fg="white").pack(side="right", padx=10)

        # Scrollable list of resources
        list_frame = tk.Frame(self.content_frame, bg="#F0F4F8", bd=1, relief="solid")
        list_frame.pack(fill="both", expand=True, padx=10, pady=10)

        canvas = tk.Canvas(list_frame, bg="#F0F4F8")
        scrollbar = tk.Scrollbar(list_frame, orient="vertical", command=canvas.yview)
        self.resource_list_frame = tk.Frame(canvas, bg="#F0F4F8")

        self.resource_list_frame.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
        canvas.create_window((0, 0), window=self.resource_list_frame, anchor="nw")
        canvas.configure(yscrollcommand=scrollbar.set)

        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")

        # initially show all resources
        self.displayed_resources = self.resources.copy()
        self.render_resource_items()

    def render_resource_items(self):
        # clear current items
        for w in self.resource_list_frame.winfo_children():
            w.destroy()

        if not self.displayed_resources:
            tk.Label(self.resource_list_frame, text="No resources found.", bg="#F0F4F8").pack(pady=20)
            return

        for res in self.displayed_resources:
            card = tk.Frame(self.resource_list_frame, bg="white", bd=1, relief="solid", padx=10, pady=8)
            card.pack(fill="x", padx=10, pady=8)

            title = tk.Label(card, text=res.get("title", "Untitled"), font=("Arial", 14, "bold"), bg="white")
            title.pack(anchor="w")

            meta = tk.Label(card, text=f"{res.get('type','')} • {res.get('category','')}", bg="white", font=("Arial", 10, "italic"))
            meta.pack(anchor="w")

            desc = tk.Label(card, text=res.get("description", ""), bg="white", wraplength=800, justify="left")
            desc.pack(anchor="w", pady=(4, 6))

            btn_frame = tk.Frame(card, bg="white")
            btn_frame.pack(anchor="e")

            open_btn = tk.Button(btn_frame, text="Open Link", command=lambda url=res.get("url"): self.open_link(url), bg="#6C63FF", fg="white")
            open_btn.pack(side="right", padx=5)

            copy_btn = tk.Button(btn_frame, text="Copy URL", command=lambda url=res.get("url"): self.copy_to_clipboard(url))
            copy_btn.pack(side="right", padx=5)

            # simple remove (admin)
            remove_btn = tk.Button(btn_frame, text="Remove", command=lambda r=res: self.remove_resource(r), bg="#FF4D4D", fg="white")
            remove_btn.pack(side="right", padx=5)

    def open_link(self, url):
        if not url:
            messagebox.showwarning("No Link", "No link is available for this resource.")
            return
        try:
            webbrowser.open(url, new=2)
        except Exception as e:
            messagebox.showerror("Error", f"Could not open link: {e}")

    def copy_to_clipboard(self, text):
        self.master.clipboard_clear()
        self.master.clipboard_append(text)
        messagebox.showinfo("Copied", "URL copied to clipboard.")

    def filter_resources(self):
        q = self.res_search_var.get().strip().lower()
        if not q:
            self.displayed_resources = self.resources.copy()
        else:
            filtered = []
            for r in self.resources:
                if q in r.get("title","").lower() or q in r.get("description","").lower() or q in r.get("category","").lower() or q in r.get("type","").lower():
                    filtered.append(r)
            self.displayed_resources = filtered
        self.render_resource_items()

    def show_all_resources(self):
        self.res_search_var.set("")
        self.displayed_resources = self.resources.copy()
        self.render_resource_items()

    def add_resource_dialog(self):
        # Basic dialog for adding a resource (admin)
        title = simpledialog.askstring("Add Resource", "Title (e.g., 'Coursera - Machine Learning'):")
        if not title:
            return
        rtype = simpledialog.askstring("Add Resource", "Type (Study Material / Online Course / University Link):", initialvalue="Online Course")
        category = simpledialog.askstring("Add Resource", "Category (e.g., 'Computer Science', 'Biology'):", initialvalue="General")
        description = simpledialog.askstring("Add Resource", "Short Description:")
        url = simpledialog.askstring("Add Resource", "URL (full link):")
        new_res = {
            "title": title,
            "type": rtype or "Online Course",
            "category": category or "General",
            "description": description or "",
            "url": url or ""
        }
        self.resources.insert(0, new_res)  # newest first
        save_resources(self.resources)
        self.show_all_resources()
        messagebox.showinfo("Added", "Resource added successfully.")

    def remove_resource(self, res):
        if messagebox.askyesno("Remove Resource", f"Remove '{res.get('title')}'?"):
            try:
                self.resources.remove(res)
                save_resources(self.resources)
            except ValueError:
                pass
            self.show_all_resources()

# --- Run App ---
if __name__ == "__main__":
    root = tk.Tk()
    app = SkillRouteAnalyzer(root)
    root.mainloop()
```
